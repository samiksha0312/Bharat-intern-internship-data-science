```python
import matplotlib.pyplot as plt
import pandas as pd

# Given dataset
data = {
    'ID': ['ff0fda5f289ae858781f9aa093df3970', 'ff1233d348cba9f95ba35679b35e26b6', 'ff18e23435cd4041290292f36014d7e3',
           'ff2512ccb9ac4ede6c0c56a8f5860c95', 'ff31c2bc7cf296899ce28a809600e20e', 'ff3410ab3513641567039f0a443d91dc',
           'ff443bbc0fdb63893b066cb7d7af4aa0', 'ff5aacc3eeb12cc9c029259d5171a4df', 'ff602c36a58c20d88fd0f543fd671040',
           'ff6123b262da1a561197a5a25c212964', 'ff6591fb1b3481fe0c366dfdcdc8ab5f', 'ff6bd913df168371937cfd31e01c058d',
           'ff7085719aa9207b7e00250264b4b63f', 'ff75490f8f99867fc9dec1c298e0761a'],
    'Animal': ['dog', 'cat', 'dog', 'dog', 'cat', 'cat', 'dog', 'dog', 'dog', 'dog', 'dog', 'cat', 'dog', 'dog']
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Counting occurrences of each animal type
animal_counts = df['Animal'].value_counts()

# Plotting
plt.figure(figsize=(8, 6))
plt.pie(animal_counts, labels=animal_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Animal Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
